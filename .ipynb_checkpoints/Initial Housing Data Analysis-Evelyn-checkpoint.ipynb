# London house price prediction by using HM Land Registry: Price Paid Data . 


Price Paid Data includes information on all property sales in England and Wales that are sold for full market value and are registered at HM Land Registry.

File include standard and additional price paid data transactions received at HM Land Registry from 1 January 1995 to the most current monthly data.

Accuracy of the data
Price paid information is not personal, but property-related information. All commercial transactions, sales that have not been lodged with HM Land Registry or sales that were not for full market value are excluded.

Now we need to filter the data only by the representative boroughs, meaning those boroughs that have the ore number of instaces or in other words those bouroughs in London that had more house selling.

```python
#londonBoroughs=df_london[(df_london['district'].isin(representativeBoroughs))]
#londonBoroughs.to_csv('londonBoroughs.csv',header=True)
londonBoroughs = pd.read_csv('londonBoroughs.csv', index_col=0)
```

Now let's convert categorical values in dummy values as well as separate date into its corresponding dd-mm-yyy fields.

```python
#convert price to numeric
londonBoroughs['price'] = pd.to_numeric(londonBoroughs['price'])
londonBoroughs['dateOfTransfer'] = pd.to_datetime(londonBoroughs['dateOfTransfer'])
londonBoroughs['year']=londonBoroughs['dateOfTransfer'].dt.year
londonBoroughs['month']=londonBoroughs['dateOfTransfer'].dt.month
londonBoroughs['day']=londonBoroughs['dateOfTransfer'].dt.day
```

```python

londonBoroughs = pd.concat([londonBoroughs,pd.get_dummies(londonBoroughs['propertyType'], prefix='prop_type')],axis=1)
londonBoroughs = pd.concat([londonBoroughs,pd.get_dummies(londonBoroughs['oldNew'], prefix='oldNew')],axis=1)
londonBoroughs = pd.concat([londonBoroughs,pd.get_dummies(londonBoroughs['duration'], prefix='duration')],axis=1)
londonBoroughs = pd.concat([londonBoroughs,pd.get_dummies(londonBoroughs['district'], prefix='district')],axis=1)
```

We need to drop columns that are not going to be used ,including date which we have refformated separately and the ones  we have converted into dummy values as well as the ones we are not going to use for prediction :  :

```python
londonBoroughs=londonBoroughs.drop(['uniqueId','postCode','PAON','SAON','street','locality','recordStatus','ppdCategoryType','dateOfTransfer','propertyType','oldNew','duration','district','townCity','county'],axis=1)
```

```python
londonBoroughs.head(3)
```

```python
corrmat = londonBoroughs.corr()
f, ax = plt.subplots(figsize=(12, 9))
sns.heatmap(corrmat, vmax=.8, square=True);
```

```python
import numpy as np
k=10
cols=corrmat.nlargest(k,'price')['price'].index
cm=np.corrcoef(londonBoroughs[cols].values.T)
sns.set(font_scale=1.25) 
hm=sns.heatmap(cm, cbar=True, annot=True, square=True, fmt='.2f',annot_kws={ 'size': 10 }, yticklabels=cols.values,xticklabels=cols.values)
plt.show()
```
